```python
def search_gene(fna_file, gene_name):
    found_gene = False
    output = ""

    with open(fna_file, 'r') as f:
        gene_info = ""
        sequence = ""
        for line in f:
            if line.startswith('>'):
                if gene_info and sequence:
                    if gene_info.startswith('>' + gene_name) or gene_info.find(gene_name) != -1:
                        output += gene_info + '\n'
                        found_gene = True
                    gene_info = ""
                    sequence = ""
                gene_info = line.strip()
            else:
                sequence += line.strip()

        if gene_info and sequence:
            if gene_info.startswith('>' + gene_name) or gene_info.find(gene_name) != -1:
                output += gene_info + '\n'
                found_gene = True

    return output if found_gene else "Gene not found in the FNA file."

# Example usage:
fna_file = "/content/drive/MyDrive/Data/output.fna"
gene_name = input("Enter the gene name you want to search ") #Enter the gene you want to search

result = search_gene(fna_file, gene_name)
print(result)
```
